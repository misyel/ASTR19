```python
import math
import numpy as np
```

```python
def cos(x): # returns cos(x)
    return math.cos(x)
```

```python
def sin(x): # returns sin(x)
    return math.sin(x)
```

```python
def calculateSinCos():
    x = np.arange(0, 2 * math.pi, math.pi / 500)
    print("Sin")
    for i in x:
        print('{0:.6f}\t{1:.6f}'.format(i, sin(i)))
    print("\nCos")
    for i in x:
        print('{0:.6f}\t{1:.6f}'.format(i, cos(i)))
```

```python
calculateSinCos()
```

```python
def firstTenSinCos():
    x = np.arange(0, 2 * math.pi, math.pi / 500)
    for i in range(0, 10):
        print('{0:.6f}\t{1:.6f}\t{1:.6f}'.format(x[i], sin(x[i]), cos(x[i])))
```

```python
firstTenSinCos()
```
